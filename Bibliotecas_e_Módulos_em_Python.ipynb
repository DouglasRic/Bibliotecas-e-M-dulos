```python
# Aula 4 ‚Äì Bibliotecas e M√≥dulos em Python

# Passo 1: Definindo a classe Livro
class Livro:
    def __init__(self, titulo, autor, genero, quantidade):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.quantidade = quantidade

    def __str__(self):
        return f'T√≠tulo: {self.titulo} | Autor: {self.autor} | G√™nero: {self.genero} | Quantidade: {self.quantidade}'


# Passo 2: Criando a lista de livros5
biblioteca = []

# Passo 3: Fun√ß√µes para gerenciar os livros

# Cadastrar novo livro
def cadastrar_livro():
    titulo = input("Digite o t√≠tulo do livro: ")
    autor = input("Digite o autor do livro: ")
    genero = input("Digite o g√™nero do livro: ")
    quantidade = int(input("Digite a quantidade dispon√≠vel: "))
    novo_livro = Livro(titulo, autor, genero, quantidade)
    biblioteca.append(novo_livro)
    print("üìó Livro cadastrado com sucesso!")

# Listar todos os livros
def listar_livros():
    if not biblioteca:
        print("Nenhum livro cadastrado.")
    else:
        print("\nüìö Lista de Livros:")
        for livro in biblioteca:
            print(livro)

# Buscar livro por t√≠tulo
def buscar_livro(titulo_procurado):
    encontrados = [livro for livro in biblioteca if livro.titulo.lower() == titulo_procurado.lower()]
    if encontrados:
        print("\nüîç Livro(s) encontrado(s):")
        for livro in encontrados:
            print(livro)
    else:
        print("üö´ Livro n√£o encontrado.")

# Passo 4: Gerar gr√°fico com Matplotlib
import matplotlib.pyplot as plt

def gerar_grafico_genero():
    if not biblioteca:
        print("Nenhum dado para gerar gr√°fico.")
        return

    generos = {}
    for livro in biblioteca:
        if livro.genero in generos:
            generos[livro.genero] += livro.quantidade
        else:
            generos[livro.genero] = livro.quantidade

    plt.figure(figsize=(10,5))
    plt.bar(generos.keys(), generos.values(), color='skyblue')
    plt.title("Quantidade de Livros por G√™nero")
    plt.xlabel("G√™nero")
    plt.ylabel("Quantidade")
    plt.grid(True)
    plt.show()

# Passo 5: Testar o sistema
# (Exemplo simples de menu interativo)

def menu():
    while True:
        print("\n--- MENU ---")
        print("1. Cadastrar Livro")
        print("2. Listar Livros")
        print("3. Buscar Livro por T√≠tulo")
        print("4. Gerar Gr√°fico por G√™nero")
        print("5. Sair")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == '1':
            cadastrar_livro()
        elif opcao == '2':
            listar_livros()
        elif opcao == '3':
            titulo = input("Digite o t√≠tulo do livro a buscar: ")
            buscar_livro(titulo)
        elif opcao == '4':
            gerar_grafico_genero()
        elif opcao == '5':
            print("Encerrando o sistema. At√© mais! üëã")
            break
        else:
            print("Op√ß√£o inv√°lida. Tente novamente.")

# Para iniciar o sistema no Google Colab, basta rodar:
menu()
```
